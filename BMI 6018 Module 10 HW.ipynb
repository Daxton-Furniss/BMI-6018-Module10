```python
import pandas as pd
```

```python
df = pd.read_csv("breast_cancer_data.csv")
df.columns = ['Class', 'Age', 'Menopause', 'Tumor-size', 'Inv-nodes', 'Node-caps', 'Deg-malig', 'Breast', 'Breast-quad', 'Irradiat']
df
```

```python
#Melt
columns_ = ['Class', 'Age', 'Menopause', 'Tumor-size', 'Inv-nodes', 'Node-caps', 'Deg-malig', 'Breast', 'Breast-quad', 'Irradiat']
x = ['Class', 'Age', 'Menopause']

melted_df = pd.melt(df, id_vars = x, var_name= 'Attributes', value_name='Values')
melted_df
```

```python
#Pivot
df1 = df.drop_duplicates(subset=['Menopause'])
df2 = df1.pivot(index= 'Menopause', columns='Age', values= 'Tumor-size')
df2
```

```python
#Aggregation 
df.agg({ 'Age' : "count", 'Inv-nodes' : "max", 'Tumor-size' : "min"})
```

```python
#Iteration
for index, row in df.iterrows():
    print(index, row['Menopause'])
```

```python
#Groupby
df.groupby('Class', as_index=False).agg({'Age' : 'count', 'Deg-malig' : 'std'})
```
